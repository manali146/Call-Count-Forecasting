Saving data into CSV, Yearwise

```python
%%time
raw_data = {}
def access_data(day, month, year, temp_raw_data):
    date = str(year)+'-'+str(month)+'-'+str(day)+'T00:00:00.000'
    daily_data = client.get("wewp-mm3p", limit=100000, date = date )                 # Extracting daily data using API
    daily_data = pd.DataFrame.from_records(daily_data)                               # Converting to DataFrame
    print('new', date, 'Rows:', daily_data.shape[0])
    temp_raw_data = pd.concat([temp_raw_data,daily_data], ignore_index=True, axis=0) # Appending year data
    return temp_raw_data

c = Calendar()

for year in range(2019, 2023):
    raw_data[year] = {}
    temp_raw_data = pd.DataFrame()
    for month in range(1,13):
        for d in [x for x in c.itermonthdates(year, month) if x.month == month]:
            temp_raw_data = access_data(d.day, d.month, year, temp_raw_data)
    temp_raw_data.to_csv('RawData_{}.csv'.format(year), sep=',') #, encoding='utf-8' # Storing yearly data into CSV
    #raw_data.update({year: temp_raw_data})                                           # Appending yearly data to one dataframe
```
